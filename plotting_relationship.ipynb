## Seaborn_Experiment

`sns.relplot` - Relating variables with scatter plots
https://seaborn.pydata.org/tutorial/relational.html

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

### 01_Generate_Data

```python
# setting the number of rows for the CSV file (1M)
N = 200

# creating a pandas dataframe (df) with 7 columns with column names from A to G
df = pd.DataFrame(
    np.random.randint(9,999, size=(N, 4)), 
    columns=list('ABCD'))

df['E'] = df['A']*2
df['F'] = df['A']*2+df['B']
df.shape
```

```python
df.head()
```

```python
# linear relationship
sns.relplot(x="A", y="E", data=df);
```

```python
# no relationship
sns.relplot(x="A", y="B", data=df);
```

```python
# linear relationship with noise
sns.relplot(x="A", y="F", data=df);
```
